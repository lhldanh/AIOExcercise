#### Ex 2: Convert to Gray Scale

```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

##### 1. Lightness method

```python
# Chuyển dog.jpeg thành ảnh xám lightness
img = mpimg.imread('D:\AIOExcercise\Module_2\\Numpy\dog.jpeg')

gray_img_01 = np.max(img, axis=2, keepdims=True) / 2 + np.min(img, axis=2, keepdims=True) / 2

print(gray_img_01[0, 0])
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Hiển thị ảnh gốc
ax1.imshow(img)
ax1.set_title('Ảnh gốc')
ax1.axis('off')

# Hiển thị ảnh xám
ax2.imshow(gray_img_01, cmap='gray')
ax2.set_title('Ảnh xám (Lightness method)')
ax2.axis('off')

plt.tight_layout()
plt.show()
```

##### 2. Average method

```python
img = mpimg.imread('D:\AIOExcercise\Module_2\\Numpy\dog.jpeg')

gray_img_02 = np.mean(img, axis=2)
print(gray_img_02[0, 0])
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Hiển thị ảnh gốc
ax1.imshow(img)
ax1.set_title('Ảnh gốc')
ax1.axis('off')

# Hiển thị ảnh xám
ax2.imshow(gray_img_02, cmap='gray')
ax2.set_title('Ảnh xám (Average method)')
ax2.axis('off')

plt.tight_layout()
plt.show()
```

##### 3. Luminosity method

```python
img = mpimg.imread('D:\AIOExcercise\Module_2\\Numpy\dog.jpeg')

gray_img_03 = np.dot(img[..., :3], [0.21, 0.72, 0.07])
print(gray_img_03[0, 0])
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Hiển thị ảnh gốc
ax1.imshow(img)
ax1.set_title('Ảnh gốc')
ax1.axis('off')

# Hiển thị ảnh xám
ax2.imshow(gray_img_03, cmap='gray')
ax2.set_title('Ảnh xám (Luminosity method)')
ax2.axis('off')

plt.tight_layout()
plt.show()
```

##### 6. Call Sales.Average is A. Create Scores = (Good, Average, Bad) Sales > A is Good, < A is Bad, = A is Average 

```python
A = sales.mean()

scores = np.full(len(sales), 'Good')

scores[sales < A] = 'Bad'
scores[sales == A] = 'Average'

scores[7:10]
```

##### 7. new_A is closet value with A. Create Scores

```python
closet_index = np.argmin(np.abs(sales - A))
new_A = sales[closet_index]

scores = np.full(len(sales), 'Good')

scores[sales < new_A] = 'Bad'
scores[sales == new_A] = 'Average'

scores[7:10]
```
